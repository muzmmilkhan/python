### Reading Data From Different Sources

```python
import pandas as pd
from io import StringIO
```

```python
Data = '{"employee_name": "John Doe", "employee_id": 12345, "department": "Engineering", "salary": 75000, "Job_Profile": "Software Engineer", "location": "New York"}'
```

```python
df = pd.read_json(StringIO(Data), lines=True)
```

```python
df
```

```python
df.to_json('employee_data.json', orient='records', lines=True)
```

```python
# Read wine_data from website
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
wine_data = pd.read_csv(url, sep=';')
```

```python
wine_data.head()
```

```python
wine_data.to_csv('wine_data.csv', index=False)
```

```python
# Read banklist from website
banklist_url = 'https://www.fdic.gov/bank/individual/failed/banklist.html'
banklist_data = pd.read_html(banklist_url)[0]
```

```python
banklist_data.head()
```

```python
# Read Mobile Country Codes from website
mcc_url = 'https://en.wikipedia.org/wiki/Mobile_country_code'
mcc_data = pd.read_html(mcc_url)[0]
mcc_data.head()
```
